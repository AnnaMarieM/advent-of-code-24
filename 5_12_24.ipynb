```python
with open("05input (1).txt", "r") as file:
#with open("05input.txt", "r") as file:
    pages = file.read().split("\n")

split_index = pages.index('')
rules = [list(map(int, row.split('|'))) for row in pages[:split_index]]
pages_produce = [list(map(int, row.split(','))) for row in pages[split_index + 1:]]
```

```python
rules_dict = {}

for rule in rules:
  if rule[0] not in rules_dict.keys():
    rules_dict[rule[0]] = set()
  rules_dict[rule[0]].add( rule[1])


def validate_order(element_list, rules):
  for i, element in enumerate(element_list):
    preceding_elements = element_list[:i]
    if element in rules and any(preceding in rules[element] for preceding in preceding_elements):
      return False
  return True


def put_in_order(element_list, rules):
    from collections import defaultdict, deque
    graph = defaultdict(set)  # adjacency list for the DAG
    in_degree = defaultdict(int)  # count of incoming edges

    for elem in element_list:
        graph[elem] = set()
        in_degree[elem] = 0

    for key, disallowed in rules.items():
        for node in disallowed:
            if node in element_list and key in element_list:
                graph[node].add(key)
                in_degree[key] += 1

    queue = deque([node for node in element_list if in_degree[node] == 0])
    ordered_list = []

    while queue:
        current = queue.popleft()
        ordered_list.append(current)
        for neighbor in graph[current]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    if len(ordered_list) == len(element_list):
        return ordered_list
    return [0]
```

```python
# Part 1 + 2
sum_of_middle_pages = 0
sum_of_incorrectly_placed_pages = 0

for tisk in pages_produce:
  if validate_order(tisk, rules_dict):
    middle_index = len(tisk) // 2
    sum_of_middle_pages += tisk[middle_index]
  else:
    tisk = put_in_order(tisk, rules_dict)
    middle_index = len(tisk) // 2
    sum_of_incorrectly_placed_pages += tisk[middle_index]

print(sum_of_middle_pages)
print(sum_of_incorrectly_placed_pages)
```
